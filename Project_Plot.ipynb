```c++
double pT[5] = {0.326, 0.752, 1.226, 1.715, 2.471};
double lambda[5] = {1.47, 0.86, 1.16, 1.00, 1.33};
double mu[5] = {0.031, -0.007, 0.013, -0.040, 0.012};
double nu[5] = {-0.024, 0.044, 0.057, 0.035, 0.031};

double e_pT_lower[5] = {0.326, 0.252, 0.226, 0.215, 0.471};
double e_pT_upper[5] = {0.179, 0.248, 0.279, 0.285, 1.529};
double e_lam[5] ={0.16, 0.11, 0.13, 0.19, 0.24};
double e_mu[5] ={0.029, 0.021, 0.026, 0.041, 0.057};
double e_nu[5] ={0.024, 0.017, 0.020, 0.030, 0.041};
```

```c++
double lambda_pp[5] = {1.32, 0.91, 0.74, 1.06, 0.59};
double mu_pp[5] = {0.097, -0.007, -0.044, -0.007, -0.041};
double nu_pp[5] = {0.019, 0.051, 0.039, 0.046, 0.195};

double e_lam_pp[5] ={0.24, 0.17, 0.19, 0.30, 0.34};
double e_mu_pp[5] ={0.045, 0.032, 0.040, 0.065, 0.087};
double e_nu_pp[5] ={0.036, 0.026, 0.031, 0.046, 0.060};
```

```c++
double lt[5], e_lt[5], lt_pp[5], e_lt_pp[5];

for (int i =0 ; i<6; i++){
    
    lt[i]=2*mu[i]-(1-lambda[i]);
    lt_pp[i]=2*mu_pp[i]-(1-lambda_pp[i]);
    e_lt[i]=sqrt(4*e_mu[i]*e_mu[i] + e_lam[i]*e_lam[i]);
    e_lt_pp[i]=sqrt(4*e_mu_pp[i]*e_mu_pp[i] + e_lam_pp[i]*e_lam_pp[i]);
    }
```

```c++
auto g_lt = new TGraphAsymmErrors(5, pT, lt, e_pT_lower, e_pT_upper, e_lt, e_lt);
auto g_nu = new TGraphAsymmErrors(5, pT, nu, e_pT_lower, e_pT_upper, e_nu, e_nu);
auto g_mu = new TGraphAsymmErrors(5, pT, mu, e_pT_lower, e_pT_upper, e_mu, e_mu);
auto g_lam = new TGraphAsymmErrors(5, pT, lambda, e_pT_lower, e_pT_upper, e_lam, e_lam);
```

```c++
auto g_lt_pp = new TGraphAsymmErrors(5, pT, lt_pp, e_pT_lower, e_pT_upper, e_lt_pp, e_lt_pp);
auto g_nu_pp = new TGraphAsymmErrors(5, pT, nu_pp, e_pT_lower, e_pT_upper, e_nu_pp, e_nu_pp);
auto g_mu_pp = new TGraphAsymmErrors(5, pT, mu_pp, e_pT_lower, e_pT_upper, e_mu_pp, e_mu_pp);
auto g_lam_pp = new TGraphAsymmErrors(5, pT, lambda_pp, e_pT_lower, e_pT_upper, e_lam_pp, e_lam_pp);
```
